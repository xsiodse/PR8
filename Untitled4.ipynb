```python
import re

# Паттерн для простейшего email: word@word.com/net/org
# (более строгие паттерны сложнее для учебной задачи)
EMAIL_PATTERN = r'[\w.-]+@[\w-]+\.(com|net|org)$'



def validate_age():
    while True:
        age_input = input("Введите ваш возраст (целое число от 0 до 120): ")
        try:
            # 1. Проверка типа данных (вызовет ValueError, если не число)
            age = int(age_input)

            # 2. Проверка диапазона (логическая проверка)
            if 0 <= age <= 120:
                print(f"✅ Возраст принят: {age}")
                return age
            else:
                print("❌ Возраст должен быть в диапазоне от 0 до 120.")
        except ValueError:
            print("❌ Ошибка ввода. Пожалуйста, введите **целое число**.")


def validate_email():
    while True:
        email = input("Введите ваш email (формат: user@domain.com): ").strip()

        # re.fullmatch проверяет, соответствует ли ВСЯ строка паттерну
        if re.fullmatch(EMAIL_PATTERN, email):
            print(f"✅ Email принят: {email}")
            return email
        else:
            print("❌ Неверный формат email. Убедитесь, что адрес содержит @ и домен .com/.net/.org")


print("--- Проверка возраста ---")
user_age = validate_age()

print("\n--- Проверка email ---")
user_email = validate_email()

print("\n--- Сводка ---")
print(f"Ввод завершен. Возраст: {user_age}, Email: {user_email}")
```

```python
import re

def validate_inn():
    # Паттерн: ^ - начало, \d{12} - ровно 12 цифр, $ - конец строки
    inn_pattern = r"^\d{12}$"

    while True:
        inn_input = input("Введите ваш ИНН (12 цифр): ").strip()

        if re.fullmatch(inn_pattern, inn_input):
            print(f"✅ ИНН принят: {inn_input}")
            return inn_input
        else:
            print("❌ Ошибка: ИНН должен содержать ровно 12 цифр.")

# Запуск
if __name__ == "__main__":
    validate_inn()
```

```python
def validate_price():
    while True:
        price_input = input("Введите цену товара (например, 499.99): ").strip()

        # Заменяем запятую на точку для поддержки обоих форматов ввода
        price_input = price_input.replace(',', '.')

        try:
            price = float(price_input)

            if 0 < price <= 1000000:
                print(f"✅ Цена принята: {price:.2f}")
                return price
            elif price <= 0:
                print("❌ Ошибка: цена должна быть больше нуля.")
            else:
                print("❌ Ошибка: цена не может превышать 1 000 000.")

        except ValueError:
            print("❌ Ошибка: введите корректное число.")

# Запуск
if __name__ == "__main__":
    validate_price()
```

```python
import re

def validate_password():
    while True:
        password = input("Придумайте пароль (8-20 символов, цифра, заглавная буква): ").strip()

        # 1. Проверка длины
        if not (8 <= len(password) <= 20):
            print("❌ Ошибка: длина пароля должна быть от 8 до 20 символов.")
            continue

        # 2. Проверка заглавной буквы
        if not any(char.isupper() for char in password):
            print("❌ Ошибка: пароль должен содержать хотя бы одну заглавную букву.")
            continue

        # 3. Проверка цифры
        if not any(char.isdigit() for char in password):
            print("❌ Ошибка: пароль должен содержать хотя бы одну цифру.")
            continue

        print("✅ Пароль принят и надежно сохранен!")
        return password

if __name__ == "__main__":
    validate_password()
```
